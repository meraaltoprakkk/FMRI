**Veri Seti Hikayesi**

```python
##VERİ SETİ HİKAYESİ : FMRI : Beyne bağlanan bir cihaz aracıyla toplanan sinyalleri ifade eden veri setidir.
####Değişkenler
#subject:verilerin toplandıgı kisiler
#timepoint: zaman noktaları
#event: birbirinden farklı olaylar
#region:sinyalin toplandığı bölgeler
#signal: sinyal aralığı

import seaborn as sns
fmri = sns.load_dataset("fmri")
df= fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
##Her bir zaman noktasında acaba eşit sinyal mi var? Her bir zaman noktasında eşit miktarda veri toplanmış?
df.groupby("timepoint")["signal"].count()
```

```python
#bu gruplama içinde genel istatistikler ; betimsel istatistiğini alalım
df.groupby("timepoint")["signal"].describe()
```

```python
#Çizgi Grafik ve Çaprazlamalar
##Zaman noktalarına göre sinyal verileri tutulmuş
sns.lineplot(x="timepoint", y ="signal", data=df);
```

```python
sns.lineplot(x="timepoint",y= "signal", hue ="event" , data=df);
```

```python
sns.lineplot(x= "timepoint" ,
 y= "signal", 
 hue ="event" , 
 style= "event" ,
 markers = True, dashes= False, data=df);
```
